**Steps**

**Import necessary libraries:** Each model requires specific imports for the model itself, preprocessing functions, and decoding predictions.

**Load the pre-trained model:** weights='imagenet' ensures you're using the pre-trained weights on the ImageNet dataset.

**Load and preprocess the image:** Load the image, convert it to a NumPy array, expand dimensions, and apply the appropriate preprocessing function for the model.

**Make predictions:** Use the predict method of the model to get predictions.

**Decode predictions:** Use the decode_predictions function to convert the predictions into human-readable labels.

**Important points:**

**Input size:** Xception requires a different input size (299x299) compared to the other models (224x224).

**Preprocessing:** Each model has its own preprocessing function (e.g., preprocess_input). Make sure to use the correct one.

**Image path:** Replace "African_Bush_Elephant.jpg" with the actual path to your image.

**Implementation using Resnet50**

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np



```

```python
# Load the pre-trained model
model = ResNet50(weights='imagenet')
```

```python
# Load and preprocess the image
img_path = "African_Bush_Elephant.jpg"
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
x = tf.keras.preprocessing.image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Make predictions
preds = model.predict(x)
```

```python
# Decode the predictions
print('Predicted:', decode_predictions(preds, top=1)[0])
```

**Implementation with Xception**

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input, decode_predictions
import numpy as np
```

```python
# Load the pre-trained model
model = Xception(weights='imagenet')
```

```python
# Load and preprocess the image
img_path = "African_Bush_Elephant.jpg"
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(299, 299)) # Note: Xception requires 299x299 input
x = tf.keras.preprocessing.image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Make predictions
preds = model.predict(x)
```

```python
# Decode the predictions
print('Predicted:', decode_predictions(preds, top=1)[0])
```
